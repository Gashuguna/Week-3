1. Loading the Data

```
#importing the relevant libraries

import numpy as np
import pandas as pd
```

```
#Loading the data

from google.colab import files
up_files = files.upload()
```

```
import io
cells_geo = pd.read_csv(io.BytesIO(up_files['cells_geo.csv']))
telcom1 = pd.read_csv(io.BytesIO(up_files['Telcom_dataset.csv']))
telcom2 = pd.read_csv(io.BytesIO(up_files['Telcom_dataset2.csv']))
telcom3 = pd.read_csv(io.BytesIO(up_files['Telcom_dataset3.csv']))
```

**Cleaning Data**

```
#Checking the null values

cells_geo.isnull().sum()
```

```
#Checking the null values

telcom1.isnull().sum()
```

```
#Dropping null values

telcom1.dropna(axis = 1, how = 'all', inplace = False)
```

```
#Checking the null values again

telcom1.isnull().sum()
```

```
#Combine the Datasets
new_data = pd.concat([telcom1,telcom2,telcom3, cells_geo], axis=0)
new_data.head(3)
```

```
#Dropping the data that is not needed in this analysis

new_data.drop(['COUNTRY_A', 'COUNTRY_B', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'PRODUCT', 'DW_A_NUMBER'], axis= 1)
new_data
```

```
#Checking for duplicates

new_data.dropna(axis= 1, how= 'all', inplace= True)
new_data
```

```
#Removing any possible duplicates

new_data.drop_duplicates()
```

```
#Checking the data again

new_data.head(3)
```
